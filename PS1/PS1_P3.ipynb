```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png','pdf')
```

```python
m,b = 5.0, -2.0

def gen_data(num):
    xdat = 1.*np.random.rand(num)
    ydat = m*xdat+b+np.random.normal(0,1.0,num)
    return np.stack((xdat,ydat),axis=-1)
```

```python
dat = gen_data(100)
plt.scatter(dat[:,0],dat[:,1])
```
