```python
%matplotlib inline
import fsps
import pylab as p
import numpy as np
```

```python
sp = fsps.StellarPopulation(imf_type=2,imf3=1.8,tage=0.01)
```

```python
spec_08 = sp.get_spectrum(tage=0.01)
sp.params["imf3"] = 2.3
spec_13 = sp.get_spectrum(tage=0.01)
sp.params["imf3"] = 2.8
spec_18 = sp.get_spectrum(tage=0.01)
```

```python
p.plot(spec_08[0],spec_08[1],label=r'$\alpha=0.8$',ls='dotted')
p.plot(spec_13[0],spec_13[1],label=r'$\alpha=1.3$',ls='dashed')
p.plot(spec_18[0],spec_18[1],label=r'$\alpha=1.8$',ls='dotted')
p.xscale('log')
p.yscale('log')
p.xlim(1e2,1e4)
p.ylim(1e-17,1e-13)
p.legend(loc=0)
p.show()
```
